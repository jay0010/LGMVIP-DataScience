## **IMPORTING  LIBRARIES**

```python
import tensorflow as tf
#tf.__version__
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Ignore the warnings
import warnings
warnings.filterwarnings('ignore')
# display all dataframe columns & rows
pd.options.display.max_columns = None
pd.options.display.max_rows = None
# to set the limit to 3 decimals
pd.options.display.float_format = '{:.7f}'.format
```

## **IMPORTING DATA**

```python
train = pd.read_csv('/content/train.csv')
train.head()
```

```python
test = pd.read_csv('/content/test.csv')
test.head()
```

## **EXPLORATORY DATA ANALYSIS & PREPROCESSING**







```python
train.describe()
```

```python
plt.figure(figsize=(10,7))
sns.countplot(train['label'])
plt.tight_layout(); 
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12,6))
sns.distplot(train['label'], axlabel="Labels", color='red')
```

## **FEATURE EXTRACTION**

```python
features = train.drop('label', axis=1)

y_train = train['label']

X_ = np.array(features)
X_train = X_.reshape(X_.shape[0], 28, 28)

X_test = np.array(test)
```

### **Using imshow() to visualise Image associated with the index number of X_train[]**




```python
plt.imshow(X_train[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

### **Verifying the correct format of data, to build and train the network**



```python
fig = plt.figure(figsize=(10,5))

for i in range(16):
  fig.add_subplot(4,4, i+1)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(X_train[i], cmap='gray')
  plt.xlabel('Digit: ' + str(y_train[i]))
  plt.tight_layout(); # to see clear graph
plt.show();
```

## **Shape of train and test set.**

```python
X_train.shape, X_test.shape
```

### **Data is not in the same shape so, need to reshape the train and test data respectively.**

```python
X_train = X_train.reshape(X_train.shape[0], 28, 28,1)
X_test = X_test.reshape(X_test.shape[0], 28, 28,1)
```

```python
X_train.shape, X_test.shape
```

### **Scale and normalize data: Divide the train and test dataset with pixel-value as an integer between 0 and 255, inclusive.**

```python
X_train.min(), X_train.max()
```

```python
X_test.min(), X_train.max()
```

```python
X_train = X_train/255.
X_test = X_test/255.
```

### **Checking min() , max()  and confirming data to be  normalized**

```python
X_train.min(), X_train.max()
```

# **Creation of ML model**

```python
# STEP 1. Create the model

model = tf.keras.Sequential([
                             tf.keras.layers.Flatten(input_shape=(28,28)),
                             tf.keras.layers.Dense(128, activation="relu"),
                             tf.keras.layers.Dense(10) #10 is the number of classes
])


# STEP 2 Compile the model

model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              optimizer=tf.keras.optimizers.Adam(),
              metrics=["accuracy"])
# STEP 3. Fit the model

history = model.fit(X_train, y_train, epochs=50)
```

# **ACCURACY AND LOSS PLOT :**

```python
pd.DataFrame(history.history).plot(figsize=(12,7))
```

# **PREDICTIONS:**

```python
predictions = np.argmax(model.predict(X_test), axis=1)
```
